```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
n_0 = [64, 128, 256, 512, 1024, 2048, 4096]
k_n_0 = [0.2250, 0.1266, 0.0533, 0.02895, 0.01223, 0.00548, 0.00286]
```

```python
def inverse(x, a, b):
    return (a / x) + b
```

```python
popt_0, pcov_0 = curve_fit(inverse, n_0, k_n_0, p0=(1,1), maxfev=10000)
```

```python
plt.plot(n_0, k_n_0)
samplerange = np.array([2**i for i in range(5, 17)])
plt.plot(samplerange, inverse(samplerange, *popt_0), label=f'Fit: a={popt_0[0]:.2f}, b={popt_0[1]:.2f}')
plt.title("Curve fit for dimension 0")
plt.legend()
```

```python
n_2 = [64, 128, 256, 512, 1024, 2048, 4096]
k_n_2 = [0.4796, 0.3340, 0.2049, 0.1426, 0.0895, 0.0592, 0.0374]
```

```python
def power(x, a, b, c):
    return a * (x**b) + c

```

```python
popt_2, pcov_2 = curve_fit(power, n_2, k_n_2, p0=(1,1,1), maxfev=10000)
```

```python
plt.plot(n_2, k_n_2)
plt.plot(samplerange, power(samplerange, *popt_2), label=f'Fit: a={popt_2[0]:.2f}, b={popt_2[1]:.2f}, c={popt_2[2]:.2f}')
```
